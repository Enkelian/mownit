## Zadanie 1

```python
from numpy import float32
from numpy import float64
from numpy import array
import matplotlib.pyplot as plt
import math
import time

N = 10000000
v = 0.53125
t = []
err = []
x = []

def countErr(k, sum):
    return abs(abs((k + 1) * v - sum) / ((k + 1) * v))

def printSummary(exact_sum, result, total_time):
    print("Blad bezwzgledny: " + str(abs(exact_sum - result)))
    print("Blad wzgledny: " + str(countErr(N-1, result)))
    print("Czas wykonania: " + str(total_time))

for i in range(0, N):
    t.append(float32(v))

exact_sum = 5312500
```

```python
def sumPrimitive(t, err, x):
    sum = 0
    for i in range(0, N):
        sum = float32(sum) + t[i]
        if (i + 1) % 25000 == 0:
            err.append(countErr(i + 1, sum))
            x.append(i + 1)
    return sum

start = time.time()
sum = sumPrimitive(t, err, x)
end = time.time()

print("Zwykle dodawanie:")
printSummary(exact_sum, sum, end - start)
```

Błąd względny jest duży, ponieważ dodając kolejno wartości v do sumy, wartość sumy coraz bardziej oddala się od wartości v. Jeżeli odległość między liczbami dodawanymi w reprezentacji float jest znaczna, mniejsza z liczb może zostać pominięta. Dzieje się tak ponieważ dodając liczby w tej reprezentacji wykładniki muszą zostać zrównane, a mantysy odpowiednio przesunięte. Przesunięcie mantysy małej liczby może prowadzić do utraty cyfr znaczących.

```python
plt.figure(figsize=(20,10))
plt.plot(x, err, ".", markersize=3)
plt.show()
```

Początkowo błąd względny pozostaje na poziomie 0, jednak po ok $10^6$ iteracjach zaczyna szybko rosnąć. Jest to wynikiem opisanego wyżej sposobu dodawania liczb w reprezentacji float.

```python
def sumRec(tab, l, r):
    if l >= r:
        return float32(tab[l])
    if abs(l - r) == 1:
        return float32(tab[l]) + float32(tab[r])

    m = l + math.floor((r - l) / 2)

    return float32(sumRec(tab, l, m - 1)) + float32(sumRec(tab, m, r))

start = time.time()
recursive_sum = sumRec(t, 0, N - 1)
end = time.time()

print("Dodawanie rekurencyjne:")
printSummary(exact_sum, recursive_sum, end - start)
```

Błąd względny w metodzie rekurencyjnej zmalał, ponieważ dodajemy w niej równe sobie lub mało odległe liczby.<br>
Czas wykonania wzrósł w stosunku do poprzedniej metody, ponieważ algorytm jest rekurencyjny - złozoność wynosi $O(nlogn)$, natomiast standardowe sumowanie ma złożoność $O(n)$.

```python
t2 = []
small = 0.0000001
big = 1000000

for i in range(0, N):
    if i % 2 == 0:
        t2.append(float32(small))
    else:
        t2.append(float32(big))

exact_result = small*(N/2) + big*(N/2)

start = time.time()
recursive_sum = sumRec(t2, 0, N-1)
stop = time.time()

print("Dodawanie rekurencyjne o niezerowym bledzie")
print("Blad bezwzgledny: " + str(abs(exact_result - recursive_sum)))
print("Blad wzgledny: " + str((exact_result - recursive_sum)/exact_result))
print("Czas wykonania: " + str(stop - start))
```

Algorytm rekurencyjny zwraca niezerowy błąd, gdy w tablicy wyjściowej liczby znacznie różniące się od sobie są na sąsiednich pozycjach.

## Zadanie 2 

```python
def kahan(t):
    sum = float32(0.0)
    err = float32(0.0)

    for i in range(0, N):
        y = float32(t[i]) - float32(err)
        temp = sum + y
        err = float32((temp - sum) - y)
        sum = temp
    return sum

start = time.time()
kahan_sum = kahan(t)
stop = time.time()
print("Algorytm Kahana:")
print("Blad bezwzgledny: " + str(abs(exact_sum - kahan_sum)))
print("Blad wzgledny: " + str((exact_sum - kahan_sum)/exact_sum))
print("Czas wykonania: " + str(stop - start))
```

Algorytm Kahana używa zmiennej err aby zniwelować błąd dodawania małej liczby do dużej. Do zmiennej err przypisywana jest wartość uzyskana następująco: od sumy powiększonej o liczbę y odejmowana jest dotychczasowa wartość sumy. Od tej różnicy (która arytmetycznie powinna być równa y) odejmowana jest liczba y. W tych dwóch krokach za każdym razem odejmowane liczby bliskie sobie, unika się w ten sposób błędów wynikających z przesunięcia cechy i mantysy. Zmienna err zawiera więc różnicę między wartością dodaną do sumy a faktyczną wartością y. W kolejnej iteracji rożnica ta jest odejmowna od kolejnego składnika sumy, który prawdopodobnie jest mniejszy od aktualnej wartości sumy. Zatem wartość utracona w poprzedniej iteracji jest odzyskiwana.

```python
start = time.time()
kahan_sum = kahan(t2)
stop = time.time()

print("Algorytm Kahana dla drugiej tablicy:")
print("Blad bezwzgledny: " + str(abs(exact_result - kahan_sum)))
print("Blad wzgledny: " + str((exact_result - kahan_sum)/exact_result))
print("Czas wykonania: " + str(stop - start))
```

Lepsze własności numeryczne algorytmu Kahana w stosunku do metody rekurencyjnej widać w przypadku drugiej tablicy, wypełnionej naprzemiennie dużymi i małymi wartościami. Błąd względny w algorytmie Kahana jest o rząd wielkości mniejszy niż w przypadkua algorytmu rekurencyjnego.

# Zadanie 3

```python
from pandas import DataFrame 
from numpy import float32
from numpy import float64
from IPython.display import HTML, display
import tabulate

def single(n_idx, s_idx, vals, asc):
    rng = range(0, N[n_idx]) if asc else range(N[n_idx] - 1, -1, -1)
    res = float32(0)
    for i in rng:
        res = res + float32(vals[s_idx][i])
    return res


def double(n_idx, s_idx, vals, asc):
    res = float64(0)
    rng = range(0, N[n_idx]) if asc else range(N[n_idx] - 1, -1, -1)
    for i in rng:
        res = res + float64(vals[s_idx][i])
    return res

def printTable(table):
    for row in table: 
        for item in row: 
            print(item, end="  ") 
        print()
        
def diffTable(table1, table2):
    res = []
    for i in range(0, 5):
        res.append([])
        for j in range(0, 5):
            res[i].append(table1[i][j]-table2[i][j])
    return res


N = [50, 100, 200, 500, 1000]
S = [2, 3.6667, 5, 7.2, 10]

DH = []
EH = []


for s_idx in range(0, 5):
    DH.append([])
    EH.append([])
    for k in range(0, N[4]):
        DH[s_idx].append(1 / (k+1) ** S[s_idx])
        EH[s_idx].append((-1) ** k * DH[s_idx][k])
        
def fillArray(sourceArr, precision, isAscending):
    resArr = []
    for s_idx in range(0, 5):
        resArr.append([])
        for n_idx in range(0, 5):
            if precision == 1:
                resArr[s_idx].append(single(n_idx, s_idx, sourceArr, isAscending))
            else:
                resArr[s_idx].append(double(n_idx, s_idx, sourceArr, isAscending))
    return resArr
```

```python
dzetaAsc1 = fillArray(DH, 1, True)
print("Dzeta dla pojedynczej prezycji, sumowanie w przód:")
display(HTML(tabulate.tabulate(dzetaAsc1)))
```

```python
dzetaAsc2 = fillArray(DH, 2, True)
print("Dzeta dla podwójnej prezycji, sumowanie w przód:")
display(HTML(tabulate.tabulate(dzetaAsc2, tablefmt='html')))
```

```python
print("Różnica między precyzjami")
display(HTML(tabulate.tabulate(diffTable(dzetaAsc2, dzetaAsc1), tablefmt='html')))
```

```python
dzetaDesc1 = fillArray(DH, 1, False)
print("Dzeta dla pojedynczej prezycji, sumowanie wstecz:")
display(HTML(tabulate.tabulate(dzetaDesc1, tablefmt='html', floatfmt=".10f")))
```

```python
dzetaDesc2 = fillArray(DH, 2, False)
print("Dzeta dla podwójnej prezycji, sumowanie wstecz:")
display(HTML(tabulate.tabulate(dzetaDesc2, tablefmt='html', floatfmt=".10f")))
```

```python
print("Różnica między precyzjami")
display(HTML(tabulate.tabulate(diffTable(dzetaDesc2, dzetaDesc1), tablefmt='html')))
```

```python
etaAsc1 = fillArray(EH, 1, True)
print("Eta dla pojedynczej prezycji, sumowanie w przód:")
display(HTML(tabulate.tabulate(etaAsc1, tablefmt='html', floatfmt=".10f")))
```

```python
etaAsc2 = fillArray(EH, 2, True)
print("Eta dla podwójnej prezycji, sumowanie w przód:")
display(HTML(tabulate.tabulate(etaAsc2, tablefmt='html', floatfmt=".10f")))
```

```python
print("Różnica między precyzjami")
display(HTML(tabulate.tabulate(diffTable(etaAsc2, etaAsc1), tablefmt='html')))
```

```python
etaDesc1 = fillArray(EH, 1, False)
print("Eta dla pojedynczej prezycji, sumowanie wstecz:")
display(HTML(tabulate.tabulate(etaDesc1, tablefmt='html', floatfmt=".10f")))
```

```python
etaDesc2 = fillArray(EH, 2, False)
print("Eta dla podwójnej prezycji, sumowanie wstecz:")
display(HTML(tabulate.tabulate(etaDesc2, tablefmt='html', floatfmt=".10f")))
```

```python
print("Różnica między precyzjami")
display(HTML(tabulate.tabulate(diffTable(dzetaDesc2, dzetaDesc1), tablefmt='html', floatfmt=".10f")))
```

Na podstawie powyższych wyników można stwierdzić, iż różnice między wyników dla podwójnej i pojedynczej prezycji są mniejsze w przypadku sumowania wstecz.<br>
Korzystając z zależności $$fl(x+y)=(x+y)(\epsilon+1)$$ można oszacować błąd:
$$\frac{fl(fl(x+y)+z)-(x+y+z)}{(x+y+z)} = \frac{(1+\epsilon)\big( (1+\epsilon)(x+y)+z\big)-(x+y+z)}{x+y+z} = $$ $$ = \epsilon + \frac{(\epsilon^2+\epsilon)(x+y)}{x+y+z}$$
Analogicznie
$$\frac{fl(x+fl(y+z))-(x+y+z)}{(x+y+z)} = \epsilon + \frac{(\epsilon^2+\epsilon)(y+z)}{x+y+z}$$ 
Zatem dla $|x+y|<|y+z|$ błąd względny jest mniejszy w przypadku dodawania w przód. W przypadku funkcji dzeta i eta dla każdych kolejnych trzech wyrazów zachodzi odwrotna zależność: |x+y|>|y+z|, zatem mniejszy błąd będzie dla dodawania wstecz.

# Zadanie 4

```python
from numpy import float32
from numpy import float64
from numpy import arange
import warnings
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

N = 1000


def xtoN32(x_0, r):
    xtab = [float32(x_0)]
    r = float32(r)
    for i in range(0, N):
        xtab.append(float32(r * xtab[i] * (1 - xtab[i])))
    return xtab


def xtoN64(x_0, r):
    xtab = [float64(x_0)]
    r = float64(r)
    for i in range(0, N):
        xtab.append(float64(r * xtab[i] * (1 - xtab[i])))
    return xtab
```

```python
def a(lastk, x_0):
    R = []
    for r in arange(1, 4, 0.0001):
        R.append(r)
        
    plt.figure(figsize=(20,10))

    for r in R:
        xtab = xtoN64(x_0, r)
        for i in (N - lastk, N):
            plt.plot(r, xtab[i], '.', markersize=2)

    plt.show()
    
a(10,0.1)
```

```python
a(10, 0.01)
```

```python
a(10,0.001)
```

Kształt powyższych wykresów nie zależy od wybranego $x_0$. 

```python
def a1(lastk):
    R = []
    for r in arange(1, 4, 0.001):
        R.append(r)
        
    plt.figure(figsize=(20,10))
    for x_0 in arange(0.1,1, 0.01):
        for r in R:
            xtab = xtoN64(x_0, r)
            for i in (N - lastk, N):
                plt.plot(r, xtab[i], '.', markersize=2)

    plt.show()
    
    
a1(10)
```

Powyższy wykres to nałożone diagramy bifurkacji dla wielu wartości $x_0$. Widać, iż od pewnych wartości r wykres rodziela się na $2^n$ gałęzi. Jednak dla większych wartości r wykres staje się chaotyczny.

```python
def b(lastk, x_0):
    R = []
    plt.subplots(figsize=(20,10))

    for r in arange(3.75, 3.8, 0.00001):
        R.append(r)
    
#     for x_0 in arange(0.1, 1, 0.1):
    for r in R:
        x32 = xtoN32(x_0, r)
        x64 = xtoN64(x_0, r)
        for k in (N-lastk, N):
            plt.subplot(1, 2, 1)
            plt.plot(r, x32[k], '.', markersize=2)
            plt.subplot(1, 2, 2)
            plt.plot(r, x64[k], '.', markersize=2)
    
    plt.show()
```

```python
b(10, 0.01)
```

```python
b(10, 0.01)
```

```python
b(10, 0.001)
```

Ponownie wykresy nie różnią się znacząco dla różnych wartości początkowych $x_0$. Różnice, które występują są spowodowane ograniczeniem reprezentacji float, jednak ogólny charakter wykresu pozostaje zachowany. Na każdym z wykresów widać charakterystyczne linie, w których skupiają się punkty. Na wykresach uzyskanych dla podwójnej prezycji linie te są nieznacznie bardziej widoczne.

```python
def c():
    r = 4
    n = 1000
    eps = 1e-4
    x_0s = []
    for x_0 in arange(0.01, 1, 0.01):
        x_0s.append(x_0)

    iters = []
    for x_0 in x_0s:
        xtab = [float32(x_0)]
        r = float32(r)
        i = 0
        while xtab[i] > eps and i < n:
            xtab.append(r * xtab[i] * (1 - xtab[i]))
            i = i + 1
        iters.append(i)
    plt.figure(figsize=(20,10))
    plt.plot(x_0s, iters, '.')
    
    plt.show()

c()
```

Liczba iteracji potrzebna do osiągnięcia 0 jest chaotyczna. Ponadto, w niektórych przypadkach ciąg nie osiąga wartości bliskiej zeru nawet dla dużej liczby iteracji, mimo że przy dokładnych obliczeniach dla każdego $x_0$ wartość ciągu $x_{n+1}=4x_n(1-x_n)$ powinna zbiegać do 0. Jest to spowodowane błędami obcięcia lub zaokrąglenia w reprezentacji pojedynczej prezycji.
